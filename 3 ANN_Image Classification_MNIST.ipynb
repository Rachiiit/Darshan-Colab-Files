```python
# Load the data
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test,y_test) = mnist.load_data()
```

```python
# We normalize x_train and x_test by dividing it by 255.0
x_train, x_test = x_train/255.0, x_test/255.0
```

We wont be converting the data into matrices of size NxD i.e. 2D arrays before passing them to the NN.

Instead, we will be using flattening to convert Nx28x28 data to Nx784, which is a 2D matrix.

```python
# Build the model

model = tf.keras.models.Sequential([
                                    tf.keras.layers.Flatten(input_shape=(28,28)),
                                    tf.keras.layers.Dense(128, activation="relu"),
                                    tf.keras.layers.Dropout(0.2),
                                    tf.keras.layers.Dense(10,activation="softmax")
])
```

```python
# Compile the model
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"]
              )
```

```python
# Train the model
r = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10)
```

```python
# Plot Loss per iteration
import matplotlib.pyplot as plt
plt.plot(r.history["loss"], label="loss")
plt.plot(r.history["val_loss"], label="val_loss")
plt.legend()
plt.show()
```

```python
# Plot accuracy per iteration
import matplotlib.pyplot as plt
plt.plot(r.history["accuracy"], label="acc")
plt.plot(r.history["val_accuracy"], label="val_acc")
plt.legend()
plt.show()
```

```python
# Evauate the model
print(model.evaluate(x_test, y_test))
```

```python
p_test = model.predict(x_test).argmax(axis=1)
# https://www.geeksforgeeks.org/numpy-argmax-python/
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, p_test)
print(cm)
```
